# **Data Processing**

## Checking for Missing Values

```python
# Check for missing values in the dataset
missing_data = titanic.isnull().sum()

# Print missing values for each column
print("Missing values in each column:")
print(missing_data)
```

## Handling Missing Values

```python
# Drop rows with missing values
titanic_cleaned1 = titanic.dropna()

# Check the new shape of the dataset after dropping rows
print("Shape after dropping rows with missing values:", titanic_cleaned1.shape)
```

### Applying Data Imputation for Missing Values

```python
titanic['age'].mean()
```

```python
# Mean Imputation
titanic['age_mean_imputed'] = titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic['age_imputed'] = titanic['age'].fillna(99)
```

```python
# Fill missing values in 'age' column with the median
titanic['age_median_imputed'] = titanic['age'].fillna(titanic['age'].median())
```

```python
# Fill missing values in 'embarked' column with the most frequent value
titanic['embarked_mode_imputed'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
```

```python
# Forward Fill
titanic['cabin_ffill'] = titanic['cabin'].fillna(method='ffill')

# Backward Fill
titanic['cabin_bfill'] = titanic['cabin'].fillna(method='bfill')
```

```python
from sklearn.preprocessing import StandardScaler

# Prepare data for KNN (need to scale the numerical columns)
numeric_features = ['pclass', 'sibsp', 'parch', 'fare', 'age']
titanic_numeric = titanic[numeric_features]

# Standardize the data
scaler = StandardScaler()
titanic_scaled = scaler.fit_transform(titanic_numeric)
titanic_scaled
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# Prepare data for KNN (need to scale the numerical columns)
numeric_features = ['pclass', 'sibsp', 'parch', 'fare', 'age']
titanic_numeric = titanic[numeric_features]

# Standardize the data
scaler = StandardScaler()
titanic_scaled = scaler.fit_transform(titanic_numeric)

# Apply KNN Imputer
imputer = KNNImputer(n_neighbors=5)
imputed_data = imputer.fit_transform(titanic_scaled)

# Add the imputed 'Age' back to the dataset
titanic['age_knn_imputed'] = imputed_data[:, numeric_features.index('age')]
```

```python
# Display a sample of the imputed data
print(titanic[['age', 'age_mean_imputed', 'age_median_imputed', 'age_knn_imputed', 'embarked', 'embarked_mode_imputed', 'cabin', 'cabin_ffill', 'cabin_bfill']].head(30))
```

```python
# Check for missing values in the dataset
titanic.isnull().sum()
```

## Handling Duplicates

```python
# Check for duplicate rows
duplicates = titanic.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicates (if any)
titanic_cleaned = titanic.drop_duplicates()

# Check the shape after removing duplicates
print("Shape after removing duplicates:", titanic_cleaned.shape)
```

## Handling Incorrect Data Types

```python
# Check the data types of each column
print(titanic.dtypes)

# Convert 'survived' and 'pclass' to category data types
titanic['survived_new'] = titanic['survived'].astype('category')
titanic['pclass_new'] = titanic['pclass'].astype('category')

# Verify the changes in data types
print(titanic.dtypes)
```

## Handling Outliers

```python
import matplotlib.pyplot as plt

# Box plot to detect outliers in 'fare' column
plt.figure(figsize=(8,6))
sns.boxplot(x=titanic['fare'])
plt.title('Box plot for Fare')
plt.show()
```

```python
# Define the upper bound for 'fare' as 95th percentile
upper_bound = titanic['fare'].quantile(0.80)

# Cap the values in 'fare' to remove extreme outliers
titanic['fare'] = np.where(titanic['fare'] > upper_bound, upper_bound, titanic['fare'])

# Verify changes by plotting the box plot again
plt.figure(figsize=(8,6))
sns.boxplot(x=titanic['fare'])
plt.title('Box plot for Fare (after capping outliers)')
plt.show()
```
